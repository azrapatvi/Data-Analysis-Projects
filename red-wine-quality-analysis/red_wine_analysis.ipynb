The dataset is imbalanced, as most samples belong to medium-quality wines(5 and 6), while low and high-quality wines are fewer.

```python
sns.boxplot(df['residual sugar'])
plt.title("Distribution of Residual Sugar Levels")
```

this defines there are outliers in residual sugar

```python
min,q1,median,q3,max=np.quantile(df['residual sugar'],[0,0.25,0.50,0.75,0.10])
print("minimum:",min)
print("q1:",q1)
print("median",median)
print("q3:",q3)
print("maximum:",max)
```

```python
iqr=q3-q1
print("IQR:",iqr)

lower_fence=q1-1.5*iqr
upper_fence=q3+1.5*iqr

print("lower fence",lower_fence) #if any value in a column is below this value then it is outlier
print("upper fence",upper_fence) # if a value in a column is greater than this then it is outlier
```

```python
def plotting(col_name,position):
    plt.subplot(4,3,position)

    sns.histplot(x=col_name,data=df,kde=True)
    plt.title(f"Distribution of {col_name}")
    
count=0
plt.figure(figsize=(14,14))
for col in df.columns:
    count+=1
    plotting(col,count)

plt.tight_layout()
plt.show()
```

```python
corr_btwn_sulp_qual=df[['sulphates','quality']].corr()
```

```python
sns.heatmap(corr_btwn_sulp_qual,annot=True)
plt.title("Correlation Between Sulphates and Wine Quality")
```

```python
# Correlation with quality
corr_with_quality=df.corr()['quality'].sort_values(ascending=False)

# Remove self-correlation: means quality
corr_no_self=corr_with_quality.drop(['quality'])

print("Correlation of Features with Wine Quality")
corr_no_self.head(1)
```

```python
print("Feature with the Weakest Correlation to Wine Quality")
corr_no_self.tail(1)
```

```python
sns.scatterplot(x='alcohol', y='quality', data=df)
plt.title("Relationship Between Alcohol and Quality")
```

```python

plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x='alcohol',
    y='sulphates',
    hue='quality',
    palette='viridis'
)

plt.title('Wine Quality by Alcohol and Sulphates')
plt.xlabel('Alcohol (%)')
plt.ylabel('Sulphates')
plt.legend(title='Quality')
plt.show()
```

```python
sns.scatterplot(
    x='volatile acidity',
    y='citric acid',
    data=df,
    hue='quality',
    size='quality',
    sizes=(20, 200),
    alpha=0.7
)
plt.title("Relationship Between Volatile Acidity and Citric Acid by Wine Quality")

```

```python
def plot_boxplot(col_name,position):
    plt.subplot(1,2,position)
    plt.title(f"{col_name} Outliers")
    sns.boxplot(df[col_name])

plt.figure(figsize=(13,6))
plot_boxplot('chlorides',1)
plot_boxplot('total sulfur dioxide',2)
```

### 6.Feature Scaling Using Standardization

```python
from sklearn.preprocessing import StandardScaler

cols=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']

scaler=StandardScaler()

df[cols] = scaler.fit_transform(df[cols])

df.head()
```
